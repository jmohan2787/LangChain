```python
from langchain.chat_models import ChatOpenAI
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from langchain.agents import Tool,initialize_agent, AgentType
from langchain.memory import ConversationBufferMemory,ConversationBufferWindowMemory,ConversationSummaryMemory
import os
from dotenv import load_dotenv
```

```python
load_dotenv(dotenv_path='.env')
```

```python
llm=ChatOpenAI(model='gpt-3.5-turbo',temperature=0)
```

```python
qa_prompt=PromptTemplate.from_template('Answer to User : {question}')
qa_chain=LLMChain(llm=llm,prompt=qa_prompt)
qa_agent=Tool(name='simple Agent', func=qa_chain.run, description='YOu need answer question to user')
```

```python
from langchain.memory import ConversationBufferMemory
```

```python
#ConversationBufferMemory
memory=ConversationBufferMemory(memory_key='chat_history',return_messages=True)
agent=initialize_agent(tools=[qa_agent],llm=llm,agent=AgentType.CONVERSATIONAL_REACT_DESCRIPTION, verbose=True,memory=memory)
```

```python
# Run multiple interactions
print("1ï¸âƒ£ First Question")
res1 = agent.run("What is LangChain?")
print("\nAnswer:", res1)

print("\n2ï¸âƒ£ Follow-up Question")
res2 = agent.run("Who created it?")
print("\nAnswer:", res2)

print("\n3ï¸âƒ£ Ask again about previous topic")
res3 = agent.run("Explain it simply again.")
print("\nAnswer:", res3)

#print(memory.chat_memory.messages)

for msg in memory.chat_memory.messages:
    print(f"{msg.type.upper()}: {msg.content}")
```
