```python
from langchain.chat_models import ChatOpenAI
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
import os
from dotenv import load_dotenv
from langchain.agents import Tool, initialize_agent, AgentType
from langchain.tools import TavilySearchResults
```

```python
load_dotenv(dotenv_path='.env')
```

```python
llm=ChatOpenAI(model='gpt-3.5-turbo', temperature=0)
```

```python
qa_prompt=PromptTemplate.from_template('Clarify the customer: {question}')
qa_llm=LLMChain(llm=llm,prompt=qa_prompt)
qa_tools=Tool(name='QA agent', func=qa_llm.run, description='Please answer customer questions')
```

```python
sum_prompt=PromptTemplate.from_template('Summerizer customer input: {text}')
sum_llm=LLMChain(llm=llm,prompt=sum_prompt)
sum_tools=Tool(name='summarizer', func=sum_llm.run, description='Summarizes long paragraphs or text content')
```

```python
web_prompt=PromptTemplate.from_template('websearch: {text}')
webllm=LLMChain(llm=llm,prompt=web_prompt)
web_tools=Tool(name='web search', func=TavilySearchResults(max_result=3).run, description='Search the internet for current and live information')
```

```python
tools=[qa_tools,sum_tools,web_tools]
```

```python
agent_executor=initialize_agent(tools=tools,llm=llm,agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,verbose=True)
```

```python
queries = [
    "What is LangGraph in LangChain?"
    "Summarize this: LangChain is a framework to build LLM apps using prompts, memory, tools, and agents.",
    "Latest news about OpenAI GPT-4o"
]
```

```python
for query in queries:
    print("\nüßë‚Äçüíª User Query:", query)
    response = agent_executor.run(query)
    print("\nü§ñ Agent Response:", response)
```
