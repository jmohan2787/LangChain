```python
from langchain.chat_models import ChatOpenAI
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
import os
from dotenv import load_dotenv
from langchain.agents import Tool
from langchain.tools import TavilySearchResults
```

```python
load_dotenv(dotenv_path='.env')
```

```python
llm=ChatOpenAI(model='gpt-3.5-turbo', temperature=0)
```

```python
qa_prompt=PromptTemplate.from_template('Clarify this customer: {question}')
qa_chain=LLMChain(llm=llm,prompt=qa_prompt)
qa_tools=Tool(name='simple QA',func=qa_chain.run,description='Please answer question which asked by customer clarly')
```

```python
summerizer_prompt=PromptTemplate.from_template('Summerize the customer input: {text}')
summerizer_chain=LLMChain(llm=llm,prompt=summerizer_prompt)
summerizer_tools=Tool(name='summerizer', func=summerizer_chain.run, description='Please summerizes the customer input: {text}')
```

```python
# search_tool=Tool(name='web search', func=TavilySearchResults(max_results=3).run, description='search from internet')
search_tool = Tool(
    name="Web Search",
    func=TavilySearchResults(max_results=3).run,
    description="Search the internet for current information"
)
```

```python
ans='what is list and tuple?'
test='Even when no failures exist, your graph returns "check_all_passed", and your mapping routes back into the same node, causing an infinite loop. Even though you marked check_all_passe as a finish node, the routing MAP overrides finish behavior, causing the loop.'
webserarch='get me stories on rich dad poor dad'
```

```python
Answer=qa_tools.run({'question': ans})
```
